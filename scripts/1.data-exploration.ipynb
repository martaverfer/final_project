<h2 style="color: #75aadb;">02 | Data Cleaning and Basic Preprocessing</h2>

### Cleaning Reviews Dataframe

```python
# For reviews_df I only want a few columns
df_clean = reviews_df[['Title', 'review/summary', 'review/text', 'review/score']]
```

```python
cl.data_analysis(df_clean)
```

```python
# Data evaluation
cl.explore_data(df_clean)
```

```python
# Delete duplicated rows
df_clean = cl.remove_duplicates(df_clean)
```

```python
# Delete nan rows in Title and review/summary columns
df_clean = cl.remove_nans_by_column(df_clean, 'Title')
df_clean = cl.remove_nans_by_column(df_clean, 'review/summary')
```

```python
df_clean.shape
```

```python
# Group reviews and aggregate
df_clean['review/text'] = df_clean['review/text'].fillna('').astype(str)
df_grouped = df_clean.groupby('Title', as_index=False).agg({
    'review/summary': ' '.join,
    'review/text': ' '.join,
    'review/score': 'mean'
})
```

```python
# Rename columns for clarity
df_grouped.rename(columns={
    'review/summary': 'all_summaries',
    'review/text': 'all_texts',
    'review/score': 'avg_score'
}, inplace=True)
```

### Cleaning Books Dataframe

```python
# üßπ Drop the columns that are not needed: links an images
books_df.drop(['image', 'previewLink', 'infoLink'], axis=1, inplace=True)
```

```python
cl.data_analysis(books_df)
```

```python
# Data evaluation
cl.explore_data(books_df)
```

### Merge dataframes

```python
# Merge the two DataFrames on 'Title'
merged_df = pd.merge(books_df, df_grouped, on='Title', how='inner')
```

```python
# Data evaluation
cl.explore_data(merged_df)
```

```python
br_df = merged_df[['Title', 'authors', 'categories','all_summaries', 'all_texts', 'avg_score']]
```

```python
# Normalize column names
br_df = cl.to_snake_case(br_df)
```

```python
# Round the average score
br_df['avg_score'] = br_df['avg_score'].round(2)
```
